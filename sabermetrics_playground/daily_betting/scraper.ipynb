```python
from bs4 import BeautifulSoup
import requests
import json
import pandas as pd
import numpy as np
import re
```

```python
url = "https://www.fangraphs.com/livescoreboard.aspx?date=2023-06-17"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
text = soup.get_text()
```

```python
text
```

```python
split_str = re.split("Today's Probable Starters and Lineups Leaderboard", text)
split_str = re.split("Contact Us - Advertise", split_str[-1])
split_str[0]
```

```python
teams = ["White Sox", "Guardians", "Tigers", "Royals", "Twins", 
         "Orioles", "Red Sox", "Yankees", "Rays", "Blue Jays",
         "Astros", "Angels", "Athletics", "Mariners", "Rangers",
         "Cubs", "Reds", "Brewers", "Pirates", "Cardinals", 
         "Braves", "Marlins", "Mets", "Phillies", "Nationals",
         "Diamondbacks", "Rockies", "Dodgers", "Padres", "Giants"]
string= split_str[0]

# Create a regex pattern to match any of the phrases in the list
pattern = r'|'.join(map(re.escape, teams))

# Iterate over the list and find all instances of the phrases
active_teams = []
for match in re.finditer(pattern, string, re.IGNORECASE):
    active_teams.append(match.group())

win_probs = re.findall(r'(\d+\.\d+|\d+)%', string)
```

```python
fg_odds = pd.DataFrame({"Team" : active_teams, "Win Prob" : win_probs})
fg_odds.head()
```
