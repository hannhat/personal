```python
import numpy as np
```

```python
data = np.array([11.8,6.7,26.8,5.0,5.4,9.7,1.6,8.6,
                 15.8,10.8,32.8,12.9,0.2,0.1,6.4,0.3,
                 15.2,3.6,18.4,2.1])
```

2(b)(i)

```python
train = data[:10]
test = data[10:]
```

2(b)(ii)

```python
# Sequential Monte Carlo functions

def draw_likelihood(par):
    return np.random.exponential(1 / par, 1)[0]

def mc(n,a,b):
    alpha = len(train) + a
    beta = sum(train) + b
    posterior_sample = np.random.gamma(alpha, 1 / beta, n)
    likelihood = np.vectorize(draw_likelihood)(posterior_sample)
    return likelihood
```

```python
# Generating and evaluating point estimates for the test set:
# Gamma(a,b) with a = 3, b = 1
a = 3
b = 1
n = 10000
mc_data = mc(10000, 1, 1)

est = np.mean(mc_data)

mse = np.mean((test - est)**2)

print("Point estimate:", est)
print("MSE", mse)
```

```python
# Generating and evaluating point estimates for the test set:
# Gamma(a,b) with a = 3, b = 1
print("95 pct credible interval: [", round(np.quantile(mc_data, 0.025),2), ", ",
       round(np.quantile(mc_data, 0.975),2), "]")
```
