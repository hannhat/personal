True Parameters: \mu = 0, \sigma^2 = 1

```python
import numpy as np
import pandas as pd
```

```python
# Generating simulated data
mu = 0
sigma = 1
N = 10000

ys = np.random.normal(mu, sigma, N)
```

```python
# Initializing prior parameters
alpha_0 = 0.001
beta_0 = 0.001
mu_0 = -1
sigma_0 = 2
omega_0 = 1

mean_sigma_0 = alpha_0/omega_0
mean_mu_0 = mu_0
mean_sq_0 = 1/sigma_0 + mu_0**2

# Sufficient stats
Sy = sum(ys)
Sy2 = sum(ys**2)
n = len(ys)

# Iterations
maxiter = 10000
delta = 0.00001
diff = 10

# One-time updates 
mu_hat = (mu_0/omega_0 + Sy)/(1/omega_0 + n)
mean_mu = mu_hat

# Prior parameters
sigma_hat = 1
alpha_hat = alpha_0 + (n + 1)/2
beta_hat = 0.001
mean_sigma = alpha_hat/beta_hat

i = 0
Thetas = np.ones((maxiter,4))
Thetas[0,] = np.array([mu_hat,sigma_hat,alpha_hat,beta_hat])
while (i < maxiter) & (diff >= delta):
    i = i + 1
    mean_sigma = alpha_hat/beta_hat
    sigma_hat = (1/omega_0**2 + n)*mean_sigma
    mean_sq_mu = 1/sigma_hat + mu_hat**2
    beta_hat = beta_0 + 0.5*(Sy2- 2*mean_mu*Sy + n*mean_sq_mu) + (1/(2*omega_0))*(mu_hat**2 - 2*mu_hat*mean_mu)

    Thetas[i,] = np.array([mu_hat,sigma_hat,alpha_hat,beta_hat])  
    diff = np.max(abs(Thetas[i,] - Thetas[i-1,]))

#summarise output

print(Thetas[i-1,])
muf = Thetas[i,0]
tauf = Thetas[i,1]
af = Thetas[i,2]
bf = Thetas[i,3]
print('Converged at ',i,' iterations')
results = np.array([[muf,af/bf,tauf],[np.mean(ys),1/np.var(ys),n/np.var(ys)]])
col = ['muf','tauf','tau muf']
ind = ['VB','MLE']
results = pd.DataFrame(results,columns = col,index=ind)
results
```
