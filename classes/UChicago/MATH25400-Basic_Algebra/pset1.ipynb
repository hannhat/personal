```python
import random
import numpy as np
import math
import itertools
```

3.1

```python
for n in range(10):
    lst = np.arange(1, n + 1)
    fixed_pts_perms = 0
    total_fixed_pts = 0
    for _ in range(100000):
        perm = lst.copy()
        np.random.shuffle(perm)
        fixed_pts_lst = lst - perm
        fixed_pts = n - np.count_nonzero(fixed_pts_lst)
        if fixed_pts == 0:
            fixed_pts_perms += 1
        total_fixed_pts += fixed_pts
    print("Percentage of random permutations with no fixed points for n=",n, ":", fixed_pts_perms / 100000)
```
