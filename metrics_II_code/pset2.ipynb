```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mean = [0, 0]
cov = [[1, 0.99], [0.99, 1]]
size = 1000

gammas = np.arange(-10,10,0.02)
stds = []
for gamma in gammas:
    results = []
    for _ in range(500):
        Q, R = np.random.multivariate_normal(mean, cov, size).T
        Vars = Q / (gamma + R)
        results.append(np.mean(Vars))
    stds.append(np.std(results))

np.random.seed()
mean = [0, 0]
cov = [[1, 0.99], [0.99, 1]]
size = 1000

gammas = np.arange(-10,10,0.02)
stds2 = []
for gamma in gammas:
    results = []
    for _ in range(500):
        Q, R = np.random.multivariate_normal(mean, cov, size).T
        Vars = Q / (gamma + R)
        results.append(np.mean(Vars))
    stds2.append(np.std(results))

plt.plot(gammas, np.log(stds), label='Simulation 1', lw=0.35)
plt.plot(gammas, np.log(stds2), label='Simulation 2', lw=0.35)
plt.xlabel("Gamma")
plt.ylabel("Log standard deviation of sample mean")
plt.show()
```
